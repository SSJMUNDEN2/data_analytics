## Namen auslesen:

```python
import csv

file_path = "orders-NaNaNaN-20231115.csv"

# Liste zum Speichern der Namen
names = []

# CSV-Datei öffnen und Daten einlesen
with open(file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)

    # Überspringe die ersten drei Zeilen (mit "=" Zeichen)
    for _ in range(3):
        next(reader)

    # Iteriere durch die Zeilen und füge Vornamen und Nachnamen zur Namen-Liste hinzu
    for row in reader:
        first_name = row[6].replace('="', '').replace('"', '').strip()  # Index für Vornamen
        last_name = row[7].replace('="', '').replace('"', '').strip()   # Index für Nachnamen
        if first_name and last_name:  # Überprüfen, ob Vornamen und Nachnamen vorhanden sind
            names.append((first_name, last_name))

# Ausgabe der Namen
for first_name, last_name in names:
    print(f"{first_name} {last_name}")
```

## Bestellung und Name

```python
import csv

file_path = "orders-NaNaNaN-20231115.csv"

# Liste zum Speichern von Namen, Telefonnummer, Adresse und Subtotal
customer_data = []

# CSV-Datei öffnen und Daten einlesen
with open(file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)

    # Überspringe die ersten drei Zeilen (mit "=" Zeichen)
    for _ in range(3):
        next(reader)

    # Iteriere durch die Zeilen und füge Namen, Telefonnummer, Adresse und Subtotal zur Kundenliste hinzu
    for row in reader:
        first_name = row[6].replace('="', '').replace('"', '').strip()  # Index für Vornamen
        last_name = row[7].replace('="', '').replace('"', '').strip()   # Index für Nachnamen

        # Korrigiere die doppelten Anführungszeichen in den Namen
        name = f"{first_name} {last_name}" if first_name and last_name else ""

        phone_number = row[8].replace('="', '').replace('"', '').strip()  # Index für Telefonnummer
        address = row[12].replace('="', '').replace('"', '').strip()  # Index für Adresse
        subtotal = float(row[27])  # Index für Subtotal

        # Überprüfen, ob alle erforderlichen Informationen vorhanden sind und die Zeile nicht den zu entfernenden Werten entspricht
        if name and phone_number and address:
            customer_data.append((name, phone_number, address, subtotal))
            print(f"{name:<50}{phone_number:<15}{address:<30}{subtotal:.2f} CHF")

# Hier können Sie den Subtotal-Wert für jede Zeile überprüfen
```
